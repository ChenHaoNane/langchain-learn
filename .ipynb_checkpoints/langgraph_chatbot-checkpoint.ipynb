```python
from dotenv import load_dotenv
load_dotenv()

from typing import Annotated, List
from typing_extensions import TypedDict
from langgraph.graph import StateGraph, START, END
from langgraph.graph.message import add_messages
from langchain_openai import ChatOpenAI
from langchain_community.tools.tavily_search import TavilySearchResults
import pandas as pd
import json
from langchain_core.messages import ToolMessage, HumanMessage, AIMessage, BaseMessage
from langgraph.checkpoint.memory import MemorySaver

# å®šä¹‰ Tavily æœç´¢å·¥å…·ï¼Œæœ€å¤§æœç´¢ç»“æœæ•°è®¾ç½®ä¸º 2
tool = TavilySearchResults(max_results=2)
tools = [tool]


class State(TypedDict):
    messages: Annotated[list, add_messages]


# å®šä¹‰ BasicToolNodeï¼Œç”¨äºæ‰§è¡Œå·¥å…·è¯·æ±‚
class BasicToolNode:
    """ä¸€ä¸ªåœ¨æœ€åä¸€æ¡ AIMessage ä¸­æ‰§è¡Œå·¥å…·è¯·æ±‚çš„èŠ‚ç‚¹ã€‚
    
    è¯¥èŠ‚ç‚¹ä¼šæ£€æŸ¥æœ€åä¸€æ¡ AI æ¶ˆæ¯ä¸­çš„å·¥å…·è°ƒç”¨è¯·æ±‚ï¼Œå¹¶ä¾æ¬¡æ‰§è¡Œè¿™äº›å·¥å…·è°ƒç”¨ã€‚
    """

    def __init__(self, tools: list) -> None:
        # tools æ˜¯ä¸€ä¸ªåŒ…å«æ‰€æœ‰å¯ç”¨å·¥å…·çš„åˆ—è¡¨ï¼Œæˆ‘ä»¬å°†å…¶è½¬åŒ–ä¸ºå­—å…¸ï¼Œ
        # é€šè¿‡å·¥å…·åç§°ï¼ˆtool.nameï¼‰æ¥è®¿é—®å…·ä½“çš„å·¥å…·
        self.tools_by_name = {tool.name: tool for tool in tools}

    def __call__(self, state: State) -> State:
        """æ‰§è¡Œå·¥å…·è°ƒç”¨
        
        å‚æ•°:
        state: åŒ…å« "messages" é”®çš„å­—å…¸ï¼Œ"messages" æ˜¯å¯¹è¯æ¶ˆæ¯çš„åˆ—è¡¨ï¼Œ
              å…¶ä¸­æœ€åä¸€æ¡æ¶ˆæ¯å¯èƒ½åŒ…å«å·¥å…·è°ƒç”¨çš„è¯·æ±‚ã€‚
        
        è¿”å›:
        æ›´æ–°åçš„çŠ¶æ€ï¼ŒåŒ…å«æ‰€æœ‰åŸå§‹æ¶ˆæ¯å’Œå·¥å…·è°ƒç”¨ç»“æœ
        """
        messages = state["messages"]
        if not messages:
            raise ValueError("è¾“å…¥ä¸­æœªæ‰¾åˆ°æ¶ˆæ¯")
        
        # è·å–æœ€åä¸€æ¡æ¶ˆæ¯ï¼ˆåº”è¯¥æ˜¯AIæ¶ˆæ¯ï¼ŒåŒ…å«å·¥å…·è°ƒç”¨ï¼‰
        last_message = messages[-1]
        
        # ç”¨äºä¿å­˜å·¥å…·è°ƒç”¨çš„ç»“æœ
        new_messages = list(messages)  # å¤åˆ¶åŸå§‹æ¶ˆæ¯åˆ—è¡¨
        
        # éå†å·¥å…·è°ƒç”¨è¯·æ±‚ï¼Œæ‰§è¡Œå·¥å…·å¹¶å°†ç»“æœæ·»åŠ åˆ°æ¶ˆæ¯å†å²
        for tool_call in last_message.tool_calls:
            # æ ¹æ®å·¥å…·åç§°æ‰¾åˆ°ç›¸åº”çš„å·¥å…·ï¼Œå¹¶è°ƒç”¨å·¥å…·çš„ invoke æ–¹æ³•æ‰§è¡Œå·¥å…·
            tool_name = tool_call["name"]
            tool_args = tool_call["args"]
            tool_id = tool_call["id"]
            
            tool_result = self.tools_by_name[tool_name].invoke(tool_args)
            
            # å°†å·¥å…·è°ƒç”¨ç»“æœä½œä¸º ToolMessage æ·»åŠ åˆ°æ¶ˆæ¯å†å²
            new_messages.append(
                ToolMessage(
                    content=json.dumps(tool_result),  # å·¥å…·è°ƒç”¨çš„ç»“æœä»¥ JSON æ ¼å¼ä¿å­˜
                    name=tool_name,  # å·¥å…·çš„åç§°
                    tool_call_id=tool_id,  # å·¥å…·è°ƒç”¨çš„å”¯ä¸€æ ‡è¯†ç¬¦
                )
            )
        
        # è¿”å›æ›´æ–°åçš„çŠ¶æ€
        return {"messages": new_messages}


# åˆ›å»ºçŠ¶æ€å›¾
graph_builder = StateGraph(State)

# åˆå§‹åŒ– OpenAI èŠå¤©æ¨¡å‹
chat_model = ChatOpenAI(model="deepseek-chat")

# å°†å·¥å…·ç»‘å®šåˆ°èŠå¤©æ¨¡å‹
llm_with_tools = chat_model.bind_tools(tools)

# å®šä¹‰èŠå¤©æœºå™¨äººèŠ‚ç‚¹å‡½æ•°
def chatbot(state: State) -> State:
    """å¤„ç†æ¶ˆæ¯å¹¶ç”Ÿæˆå›å¤"""
    messages = state["messages"]
    
    # ä½¿ç”¨ LLM å¤„ç†æ¶ˆæ¯å¹¶ç”Ÿæˆå›å¤
    response = llm_with_tools.invoke(messages)
    
    # å°† AI å›å¤æ·»åŠ åˆ°æ¶ˆæ¯åˆ—è¡¨
    new_messages = list(messages)
    new_messages.append(response)
    
    return {"messages": new_messages}

# æ·»åŠ èŠå¤©æœºå™¨äººèŠ‚ç‚¹
graph_builder.add_node("chatbot", chatbot)

# åˆ›å»ºå·¥å…·èŠ‚ç‚¹å®ä¾‹å¹¶æ·»åŠ åˆ°å›¾ä¸­
tool_node = BasicToolNode(tools=tools)
graph_builder.add_node("tools", tool_node)

from typing import Literal

# å®šä¹‰è·¯ç”±å‡½æ•°ï¼Œæ£€æŸ¥å·¥å…·è°ƒç”¨
def route_tools(state: State) -> Literal["tools", "__end__"]:
    """
    ä½¿ç”¨æ¡ä»¶è¾¹æ¥æ£€æŸ¥æœ€åä¸€æ¡æ¶ˆæ¯ä¸­æ˜¯å¦æœ‰å·¥å…·è°ƒç”¨ã€‚
    
    å‚æ•°:
    state: çŠ¶æ€å­—å…¸ï¼Œç”¨äºå­˜å‚¨å½“å‰å¯¹è¯çš„çŠ¶æ€å’Œæ¶ˆæ¯ã€‚
    
    è¿”å›:
    å¦‚æœæœ€åä¸€æ¡æ¶ˆæ¯åŒ…å«å·¥å…·è°ƒç”¨ï¼Œè¿”å› "tools" èŠ‚ç‚¹ï¼Œè¡¨ç¤ºéœ€è¦æ‰§è¡Œå·¥å…·è°ƒç”¨ï¼›
    å¦åˆ™è¿”å› "__end__"ï¼Œè¡¨ç¤ºç›´æ¥ç»“æŸæµç¨‹ã€‚
    """
    messages = state["messages"]
    if not messages:
        return "__end__"
    
    # è·å–æœ€åä¸€æ¡æ¶ˆæ¯
    last_message = messages[-1]
    
    # æ£€æŸ¥æœ€åä¸€æ¡æ¶ˆæ¯æ˜¯å¦æœ‰å·¥å…·è°ƒç”¨è¯·æ±‚
    if (isinstance(last_message, AIMessage) and 
        hasattr(last_message, "tool_calls") and 
        last_message.tool_calls):
        return "tools"  # å¦‚æœæœ‰å·¥å…·è°ƒç”¨è¯·æ±‚ï¼Œè¿”å› "tools" èŠ‚ç‚¹
    
    return "__end__"  # å¦åˆ™è¿”å› "__end__"ï¼Œæµç¨‹ç»“æŸ

# æ„å»ºå›¾çš„è¾¹
graph_builder.add_edge(START, "chatbot")

# æ·»åŠ æ¡ä»¶è¾¹ï¼Œåˆ¤æ–­æ˜¯å¦éœ€è¦è°ƒç”¨å·¥å…·
graph_builder.add_conditional_edges(
    "chatbot",  # ä»èŠå¤©æœºå™¨äººèŠ‚ç‚¹å¼€å§‹
    route_tools,  # è·¯ç”±å‡½æ•°ï¼Œå†³å®šä¸‹ä¸€ä¸ªèŠ‚ç‚¹
    {
        "tools": "tools",  # å¦‚æœéœ€è¦å·¥å…·è°ƒç”¨ï¼Œè½¬åˆ°å·¥å…·èŠ‚ç‚¹
        "__end__": END     # å¦‚æœä¸éœ€è¦å·¥å…·è°ƒç”¨ï¼Œç»“æŸå½“å‰æµç¨‹
    }
)

# å½“å·¥å…·è°ƒç”¨å®Œæˆåï¼Œè¿”å›åˆ°èŠå¤©æœºå™¨äººèŠ‚ç‚¹ä»¥ç»§ç»­å¯¹è¯
graph_builder.add_edge("tools", "chatbot")

# åˆ›å»ºå†…å­˜æ£€æŸ¥ç‚¹
memory = MemorySaver()

# ç¼–è¯‘å›¾
graph = graph_builder.compile(checkpointer=memory)

# å¯é€‰ï¼šç”Ÿæˆå¯è§†åŒ–å›¾è¡¨
graph.get_graph().draw_mermaid_png(output_file_path="chatbot.png")

# ä¸»å¾ªç¯
# while True:
#     # è·å–ç”¨æˆ·è¾“å…¥
#     user_input = input("User: ")
    
#     # å¯ä»¥éšæ—¶é€šè¿‡è¾“å…¥ "quit"ã€"exit" æˆ– "q" é€€å‡ºèŠå¤©å¾ªç¯
#     if user_input.lower() in ["quit", "exit", "q"]:
#         print("Goodbye!")  # æ‰“å°å‘Šåˆ«ä¿¡æ¯
#         break  # ç»“æŸå¾ªç¯ï¼Œé€€å‡ºèŠå¤©

#     # åˆ›å»ºåŒ…å«ç”¨æˆ·æ¶ˆæ¯çš„åˆå§‹çŠ¶æ€
#     human_message = HumanMessage(content=user_input)
#     initial_state = {"messages": [human_message]}
    
#     # ä½¿ç”¨ graph.stream å¤„ç†ç”¨æˆ·è¾“å…¥ï¼Œè·å–èŠå¤©æœºå™¨äººçš„å›å¤
#     for event in graph.stream(initial_state):
#         for value in event.values():
#             # è·å–æœ€æ–°æ¶ˆæ¯
#             messages = value.get("messages", [])
#             if messages and isinstance(messages[-1], AIMessage):
#                 # æ‰“å° AI æ¶ˆæ¯å†…å®¹
#                 print("Assistant:", messages[-1].content)

config = {"configurable": {"thread_id": "1"}}

# ç”¨æˆ·è¾“å…¥çš„æ¶ˆæ¯
user_input = "Hi there! My name is Peng."

# ç¬¬äºŒä¸ªå‚æ•° config ç”¨äºè®¾ç½®å¯¹è¯çº¿ç¨‹ ID
# åœ¨è¿™é‡Œï¼Œ"thread_id" æ˜¯å”¯ä¸€æ ‡è¯†ç¬¦ï¼Œç”¨äºä¿å­˜å’ŒåŒºåˆ†å¯¹è¯çº¿ç¨‹ã€‚
# æ¯ä¸ªå¯¹è¯çº¿ç¨‹çš„çŠ¶æ€å°†ç”± MemorySaver ä¿å­˜ä¸‹æ¥ï¼Œå› æ­¤å¯ä»¥è·¨å¤šè½®å¯¹è¯ç»§ç»­è¿›è¡Œã€‚
events = graph.stream(
    {"messages": [HumanMessage(content=user_input)]},  # ç¬¬ä¸€ä¸ªå‚æ•°ä¼ å…¥ç”¨æˆ·çš„è¾“å…¥æ¶ˆæ¯ï¼Œæ¶ˆæ¯æ ¼å¼ä¸º ("user", "è¾“å…¥å†…å®¹")
    config,  # ç¬¬äºŒä¸ªå‚æ•°ç”¨äºæŒ‡å®šçº¿ç¨‹é…ç½®ï¼ŒåŒ…å«çº¿ç¨‹ ID
    stream_mode="values"  # stream_mode è®¾ç½®ä¸º "values"ï¼Œè¡¨ç¤ºè¿”å›æµå¼æ•°æ®çš„å€¼
)

# éå†æ¯ä¸ªäº‹ä»¶ï¼Œå¹¶æ‰“å°æœ€åä¸€æ¡æ¶ˆæ¯çš„å†…å®¹
for event in events:
    # é€šè¿‡ pretty_print æ‰“å°æœ€åä¸€æ¡æ¶ˆæ¯çš„å†…å®¹
    event["messages"][-1].pretty_print()

user_input = "Remember my name?"

events = graph.stream(
    {"messages": [HumanMessage(content=user_input)]},
    config,
    stream_mode="values"
)

for event in events:
    event["messages"][-1].pretty_print()

snapshot = graph.get_state(config)

messages = snapshot.values['messages']


df = pd.DataFrame([{
    'content': msg.content,
    'message_id': msg.id,
    'type': type(msg).__name__,
    'token_usage': msg.response_metadata.get('token_usage') if hasattr(msg, 'response_metadata') else None
} for msg in messages])

df.to_csv('messages.csv', index=False)
```
